1. Create scatter plots for the following data, make sure all the plots appear in the same
plane with x-labels and y-labels evenly spaced.

a. Create scatter plots with between (mpg-hp, mpg-weight), and (hp-mpg,
hp-time-to-60) with respect to each brand.

```python
fig = plt.subplots(2,2, figsize =(20,10))    #(row,columns, figsize=(width,height))
plt.subplot(2,2,1)   #(rows, columns, first plot)
sns.scatterplot(x=df['mpg'], y=df['hp'],data = df,marker='o',s= 100)

plt.subplot(2,2,2)   #(rows, columns, second plot)
sns.scatterplot(x=df['mpg'], y=df['weightlbs'],data = df)

plt.subplot(2,2,3)   #(rows, columns, third plot)
sns.scatterplot(x=df['hp'], y=df['mpg'],data = df)

plt.subplot(2,2,4)   #(rows, columns, fourth plot)
sns.scatterplot(x=df['hp'], y=df['time-to-60'],data = df)
plt.title("hp vs time-to-60")
```

```python
fig = plt.subplots(2,2, figsize =(20,10))    #(row,columns, figsize=(width,height))
plt.subplot(2,2,1)   #(rows, columns, first plot)
sns.scatterplot(x=df['mpg'], y=df['hp'],data = df,marker='o',s= 50, hue=df['brand'])

plt.subplot(2,2,2)   #(rows, columns, second plot)
sns.scatterplot(x=df['mpg'], y=df['weightlbs'],data = df, hue=df['brand'])

plt.subplot(2,2,3)   #(rows, columns, third plot)
sns.scatterplot(x=df['hp'], y=df['mpg'],data = df, hue=df['brand'])

plt.subplot(2,2,4)   #(rows, columns, fourth plot)
sns.scatterplot(x=df['hp'], y=df['time-to-60'],data = df, hue=df['brand'])
plt.title("hp vs time-to-60")
```
